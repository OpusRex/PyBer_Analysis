```python
#Add Matplotlib inline magic command
%matplotlib inline
# DEpendencies and setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
#Read the city data file and store iit in a pandas Dataframe
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
driver_count_df = city_data_df.groupby('type')[['driver_count']].sum()
driver_count_df.head()
```

```python
#Read the ride data file and store it in a pandas Dataframe.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
#combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

#display the dataframe
pyber_data_df.head()
```

```python
#Total number  of drivers for each type of city
driver_count_df = city_data_df.groupby('type')[['driver_count']].sum()
driver_count_df.head()
```

```python
#Rename driver count column to Total Drivers
driver_count_df.rename(columns={'driver_count':'Total Drivers'}, inplace=True)
driver_count_df
```

```python
#Total number of rides by type
ride_count_df = pyber_data_df.groupby('type')[['ride_id']].count()
ride_count_df
```

```python
#Rename ride_id column(counts) to Total Rides
ride_count_df.rename(columns={'ride_id':'Total Rides'}, inplace=True)
ride_count_df
```

```python
#Get total fares by city type
fare_totals_df = pyber_data_df.groupby('type')[['fare']].sum()
fare_totals_df
```

```python
#Rename fare column to Total Fares
fare_totals_df.rename(columns={'fare':'Total Fares'}, inplace=True)
fare_totals_df
```

```python
#merge ride count and driver count into single dataframe
pyber_summary_df = pd.merge(ride_count_df, driver_count_df, on='type')
pyber_summary_df
```

```python
#Add Fare totals to the new dataframe
pyber_summary_df =  pd.merge(pyber_summary_df,  fare_totals_df, on='type')
pyber_summary_df
```

```python
#Create the calculated columns in the new dataframe

#Average Fare per Ride
pyber_summary_df['Average Fare per Ride'] = (pyber_summary_df['Total Fares']/pyber_summary_df['Total Rides'])

#Average Fare per Driver
pyber_summary_df['Average Fare per Driver'] = (pyber_summary_df['Total Fares']/pyber_summary_df['Total Drivers'])
pyber_summary_df
```

```python
#Format the columns so they show as currency
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].apply(lambda x: "${:.2f}".format(x))
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].apply(lambda x: "${:.2f}".format(x))
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].apply(lambda x: "${:.2f}".format(x))
```
